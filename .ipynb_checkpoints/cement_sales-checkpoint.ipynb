```python
import pandas as pd

# read file containing the villages/cities/town zip codes witn latitud and longitud
csv_file_path = 'villages_zc.csv'
data = pd.read_csv(csv_file_path, encoding='latin1')
data_filtered = data[['postal code', 'latitude', 'longitude']]
data_filtered
```

```python
from math import radians, sin, cos, sqrt, atan2
import pandas as pd

def haversine_single_point(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Radius of the Earth in kilometers (mean value)
    radius_earth = 6371.0

    # Calculate the distance
    distance = radius_earth * c

    return distance

def calculate_distances(lat1, lon1, lats, lons):
    distances = []
    for lat2, lon2 in zip(lats, lons):
        distance = haversine_single_point(lat1, lon1, lat2, lon2)
        distances.append(distance)
    return distances

# Example usage:
fixed_point = {'latitude': -36.9, 'longitude': -60.2833}

# Assuming 'data_filtered' is your DataFrame with columns 'postal code', 'latitude', and 'longitude'
# Replace the column names with the actual ones in your DataFrame
zipcodes_filtered = zipcodes[['postal code', 'latitude', 'longitude']]

# Calculate distances
distances = calculate_distances(fixed_point['latitude'], fixed_point['longitude'], data_filtered['latitude'], data_filtered['longitude'])

# Add distances to the DataFrame
zipcodes_filtered['distance_to_fixed_point'] = distances

# Display the DataFrame with selected columns
print(data_filtered[['postal code', 'latitude', 'longitude', 'distance_to_fixed_point']])
```

```python
import pandas as pd
import requests

def get_distance(api_key, origin, destination):
    base_url = "https://maps.googleapis.com/maps/api/directions/json"
    
    params = {
        "origin": origin,
        "destination": destination,
        "key": api_key,
    }

    response = requests.get(base_url, params=params)
    data = response.json()

    if data["status"] == "OK":
        # Extract the distance information
        distance_meters = data["routes"][0]["legs"][0]["distance"]["value"]

        # Convert distance from meters to kilometers
        distance_kilometers = distance_meters / 1000.0

        return distance_kilometers

    elif data["status"] == "ZERO_RESULTS":
        return None

    else:
        return None

# Example usage:
api_key = 'AIzaSyAmGkDcszHBSlp55WBEpI_1dDSVnDGqRDc'
origin = "-36.9,-60.2833"

# Assuming data_filtered is your DataFrame
data_filtered = data[['postal code', 'latitude', 'longitude']]

# Calculate distances for each row in data_filtered
distances = []
for index, row in data_filtered.iterrows():
    destination = f"{row['latitude']},{row['longitude']}"
    distance = get_distance(api_key, origin, destination)
    
    if distance is not None:
        distances.append(distance)
    else:
        distances.append(None)

# Add distances to the DataFrame
data_filtered['distance_km'] = distances

# Print the result
print(data_filtered[['postal code', 'latitude', 'longitude', 'distance_km']])
```
